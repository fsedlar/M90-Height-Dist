```python
import csv
with open("winequality-red.csv", 'r') as f:
    wines = list(csv.reader(f, delimiter=";"))
    qualities = [float(item[-1]) for item in wines[1:]]

    ans1 = sum(qualities) / len(qualities)
    print(ans1)
```
