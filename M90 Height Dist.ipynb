```python
#Austria
AT_Bld_Count_1 = AT_Height["1Floor"].sum()
AT_Bld_Count_2 = AT_Height["2Floors"].sum()
AT_Bld_Count_3 = AT_Height["3Floors"].sum()
AT_Bld_Count_4 = AT_Height["4Floors"].sum()
AT_Bld_Count_5 = AT_Height["5Floors"].sum()
AT_Bld_Count_6to10 = AT_Height["6to10Floors"].sum()
AT_Bld_Count_11Plus = AT_Height["11PlusFloors"].sum()

AT_Low_Rise = sum([AT_Bld_Count_1, AT_Bld_Count_2, AT_Bld_Count_3])
AT_Mid_Rise = sum([AT_Bld_Count_4, AT_Bld_Count_5, AT_Bld_Count_6to10])
AT_High_Rise = AT_Bld_Count_11Plus

AT_Total = sum([AT_Low_Rise, AT_Mid_Rise, AT_High_Rise])

print("Austria Low Rise Percentage:", round(AT_Low_Rise/AT_Total,3))
print("Austria Mid Rise Percentage:", round(AT_Mid_Rise/AT_Total,3))
print("Austria High Rise Percentage:", round(AT_High_Rise/AT_Total,3))
```
